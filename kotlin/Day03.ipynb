```kotlin
var input = """467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598..""".lines()


val symbols: List<Char> = listOf('*', '#', '$', '+');


val result = input.mapIndexed({ row, line ->
    line.mapIndexed({ col, c -> 
        if (symbols.contains(c)) {
            Pair(col, row)
        } else {
            null
        }
    }).filterNotNull()
    
}).filter {it.isNotEmpty()}.flatten()

result


```
