```kotlin
import aoc2023.ColorCount

val x = "Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green"

var handsStr = x.split(":").get(1).split(";");

handsStr.map { s ->
    s.split(",").map {
        val result = "(\\d+)\\s([a-z]+)".toRegex().find(it)?.groupValues?.orEmpty()
        ColorCount(result?.get(1)?.toIntOrNull() ?: 0, result?.get(2) ?: "na")
    }
}
```

```kotlin
val x = mapOf("foo" to 3, "bar" to 4)

x.entries.all {
    it.value > 5;
}
```

```kotlin
val hands: List<Map<String, Int>> = listOf(
    mapOf("red" to 3, "blue" to 4),
    mapOf("red" to 1, "blue" to 5),
    mapOf("red" to 1, "green" to 99)
)

hands.fold(mapOf<String, Int>()) { acc, hand ->
    val more = hand.filter { (key, value) ->
        value > acc.get(key) ?: 0
    }
    acc + more
}

```
